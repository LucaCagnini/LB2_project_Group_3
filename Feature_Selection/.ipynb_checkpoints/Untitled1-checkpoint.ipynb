```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis, ProtParamData
```

```python
''' This function returns the features (max score and average score) of a chosen scale for a specific sequence. You can use a personalized 
dictionary of scores or a one inside ProtParamData. You can check all the ProtParamData scales using the command dir(ProtParamData)'''

def get_features (seq, scale ,window): 
    sequence = ProteinAnalysis(seq) 
    if type(scale)==dict:   
        scale_result = sequence.protein_scale(scale, window)
    else:
        param_dict = getattr(ProtParamData, scale)
        scale_result = sequence.protein_scale(param_dict, window)
    print(scale_result)



```
