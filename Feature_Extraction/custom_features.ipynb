```python
#function that computes the frequence of aromatic residues in a sequence

def aromatic_residue_freq(sequence):
    aromatic=0
    aromatic_list = "HYFW"
    for aa in sequence:
        if aa in aromatic_list:
            aromatic += 1
    return aromatic / len(sequence)       
```
